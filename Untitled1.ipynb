```python
mecab=MeCab.Tagger(ipadic.MECAB_ARGS)
```

```python
def normalization(text):
  text = text.replace('\n','').replace('\r','')
  text = re.sub(r'http?://[\w/:%#\$&\?\(\)~\.=\+\-]+', '', text)
  text = re.sub(r'https?://[\w/:%#\$&\?\(\)~\.=\+\-]+','', text)
  text = re.sub(r'\d+', '0', text)
  text = text.lower()
  return neologdn.normalize(text)

def get_base_forms(text):
    # mecab = MeCab.Tagger()
    nodes = mecab.parseToNode(text)
    base_forms = []
    while nodes:
        if nodes.surface != "":
            features = nodes.feature.split(',')
            pos=features[0]
            lemma=features[6] if len(features) > 7 else nodes.surface
            if pos not in ['助詞','助動詞','記号'] and lemma not in stop_words and lemma !='BOS/EOS':
               if not lemma.isdigit():
                base_forms.append(lemma)
        nodes = nodes.next
    return base_forms


# text = "私は日本語を勉強しています。"
# text=normalization(text)
# base_forms = get_base_forms(text)
# print(base_forms)

```

```python
nlp = spacy.load('ja_core_news_sm')
def split_sentences(text):
    doc = nlp(text)
    return [sent.text.strip() for sent in doc.sents]
```

```python
def preprocess(df,col_name):
  normalized_txt=[]
  splitted_sentences=[]
  token_lists=[]

  for text in df[col_name]:
    try:
      norm=normalization(text)
      sents=split_sentences(norm)
      tokenized_sents=[get_base_forms(s) for s in sents]

      normalized_txt.append(norm)
      splitted_sentences.append(sents)
      token_lists.append(tokenized_sents)
    except Exception as e:
      print(f"[ERROR] {e}")
      normalized_txt.append('')
      splitted_sentences.append([])
      token_lists.append([])

  df['Normalized_Text']=normalized_txt
  df['Sentences']=splitted_sentences
  df['Tokenized_Sentence']=token_lists
  return df

def chunk_sentence(sentences,chunk_size=5):
  return [sentences[i:i+chunk_size] for i in range(0,len(sentences),chunk_size)]
```

```python
model_name = "christian-phu/bert-finetuned-japanese-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

labels = ["negative", "neutral", "positive"]
def analyze_sentiment_chunked(sentences, chunk_size=5):
    chunks = chunk_sentence(sentences, chunk_size)
    score_acc = {"negative": 0.0, "neutral": 0.0, "positive": 0.0}
    count = {"negative": 0, "neutral": 0, "positive": 0}

    for chunk in chunks:
        chunk_text = "".join(chunk)

        # Tokenize for the model (manually)
        inputs = tokenizer(chunk_text, return_tensors="pt", truncation=True, padding=True, max_length=512)
        with torch.no_grad():
            outputs = model(**inputs)
            probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
            pred_idx = torch.argmax(probs)
            label = labels[pred_idx]
            confidence = probs[0][pred_idx].item()

            # Accumulate scores
            score_acc[label] += confidence
            count[label] += 1

    if sum(count.values()) == 0:
        return "error", 0.0

    final_label = max(score_acc, key=score_acc.get)
    final_score = score_acc[final_label] / max(count[final_label], 1)

    return final_label, final_score
```

```python
from collections import Counter
from itertools import tee, islice
import matplotlib.pyplot as plt
import japanize_matplotlib
from matplotlib import rcParams

# Set Japanese font
rcParams['font.family'] = 'IPAexGothic'

# Helper function to create n-grams from a list of tokens
def generate_ngrams(tokens, n):
    if len(tokens) < n:
        return []
    return zip(*(islice(seq, index, None) for index, seq in enumerate(tee(tokens, n))))

# Collect all n-grams across your corpus
def collect_ngrams(df, n=2):
    all_ngrams = []
    for sentence_list in df['Tokenized_Sentence']:
        for tokens in sentence_list:
            ngrams = generate_ngrams(tokens, n)
            all_ngrams.extend([" ".join(ng) for ng in ngrams])
    return Counter(all_ngrams)
```

```python
def plot_top_ngrams(ngram_counter, top_n=20, title="Top N-grams"):
    common_ngrams = ngram_counter.most_common(top_n)
    ngrams, counts = zip(*common_ngrams)

    plt.figure(figsize=(12, 6))
    plt.barh(ngrams[::-1], counts[::-1])  # Reverse for descending order
    plt.title(title, fontsize=16)
    plt.xlabel("Frequency", fontsize=12)
    plt.ylabel(f"{len(ngrams[0].split())}-gram", fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
# Bigram example
bigram_counter = collect_ngrams(df_txt, n=2)
plot_top_ngrams(bigram_counter, top_n=30, title="Top Bigrams")

# Trigram example
trigram_counter = collect_ngrams(df_txt, n=3)
plot_top_ngrams(trigram_counter, top_n=20, title="Top Trigrams")
```

```python
from wordcloud import WordCloud
# import matplotlib.pyplot as plt
# from collections import Counter
# from matplotlib import rcParams

# Set Japanese font
# rcParams['font.family'] = 'IPAexGothic'


# Collect words into a flat list
def collect_words(df):
    all_words = []
    for sentence_list in df['Tokenized_Sentence']:
        for tokens in sentence_list:
            all_words.extend(tokens)
    return all_words

# Generate word cloud
def plot_wordcloud(words, title="Word Cloud"):
    word_freq = Counter(words)

    wc = WordCloud(
        width=800,
        height=600,
        background_color="white",
        font_path="/usr/share/fonts/truetype/ipafont-gothic/ipagp.ttf",  # Change path depending on your OS
        collocations=False,  # Avoid joining unrelated words
    ).generate_from_frequencies(word_freq)

    plt.figure(figsize=(10, 8))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(title, fontsize=16)
    plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
from itertools import combinations

# 1. Extract all co-occurring pairs
def extract_cooccurrence(tokenized_sentences, min_freq=2):
    cooccurrence_counter = Counter()

    for sentence_list in tokenized_sentences:
        for tokens in sentence_list:
            tokens = list(set(tokens))  # remove duplicates within sentence
            if len(tokens) >= 2:
                for pair in combinations(tokens, 2):
                    pair = tuple(sorted(pair))  # order the pair alphabetically
                    cooccurrence_counter[pair] += 1

    # Filter by minimum frequency
    return {pair: freq for pair, freq in cooccurrence_counter.items() if freq >= min_freq}

# 2. Build the graph
def build_cooccurrence_graph(cooccurrence_dict):
    G = nx.Graph()
    for (w1, w2), weight in cooccurrence_dict.items():
        G.add_edge(w1, w2, weight=weight)
    return G

# 3. Visualize the graph
def plot_cooccurrence_graph(G, title="Co-occurrence Network"):
    pos = nx.spring_layout(G, k=0.5, iterations=50)

    plt.figure(figsize=(12, 8))
    nx.draw_networkx_nodes(G, pos, node_size=500, node_color="lightblue")
    nx.draw_networkx_edges(G, pos, width=[d['weight']*0.2 for (u,v,d) in G.edges(data=True)])
    nx.draw_networkx_labels(G, pos, font_size=10, font_family="IPAexGothic")  # For Japanese labels
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
cooccurrence = extract_cooccurrence(df_txt['Tokenized_Sentence'].iloc[0], min_freq=3)
G = build_cooccurrence_graph(cooccurrence)
plot_cooccurrence_graph(G, title="Co-occurrence of Terms")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
# from matplotlib import rcParams

# Set Japanese font
rcParams['font.family'] = 'IPAexGothic'
japanize_matplotlib.japanize()
# Build your graph G here (as before)

# Compute weighted degree for each node
weighted_degree = dict(G.degree(weight='weight'))

# Set node sizes based on weighted degree
node_sizes = [weighted_degree[node]*50 for node in G.nodes()]  # scale factor to adjust size

# Set edge width based on co-occurrence weight
edge_widths = [d['weight']*0.2 for (u, v, d) in G.edges(data=True)]

# Draw graph
pos = nx.spring_layout(G, k=0.5, iterations=50)
plt.figure(figsize=(14, 10))

nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color="lightblue")
nx.draw_networkx_edges(G, pos, width=edge_widths, alpha=0.7)
nx.draw_networkx_labels(G, pos, font_size=10)


plt.title("Co-occurrence Network with Node Sizes (Weighted Degree)", fontsize=16)
plt.axis('off')
plt.show()
```
